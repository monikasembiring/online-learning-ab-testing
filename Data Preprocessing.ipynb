# Data Preprocessing

```python
# How many videos? - 23
len(np.unique(df_videos.video_id))
```

```python
# How many distinct participants? - 254
len(np.unique(df_part.participant_id))
```

```python
# How many people acutally record their videos - 73
len(df_part[df_part['duration_recording'].isna() == False].participant_id)
```

```python
# drop unused columns from participant data
df_part_useful = df_part[['participant_id', 'video_id', 'recording_elegible', 'cam_allowed',
                         'duration_recording', 'survey_duration', 'demo_gender', 'demo_race', 
                         'correct_pre_quiz', 'correct_post_quiz']]
```

```python
# add a new column for heterogeneous effect 
df_part_useful['pre_quiz_all_correct'] = df_part_useful['correct_pre_quiz'].apply(lambda x: 1 if x == 2 else 0)
```

```python
# if a participant take the experiment more than once, only keep the first record
df_part_useful = df_part_useful.drop_duplicates(subset = 'participant_id', keep = 'first')
df_part_useful = df_part_useful.reset_index(drop = True)
```

```python
# drop unused columns from participant video data
df_part_vi_useful = df_part_vi[['participant_id', 'video_id', 'frame_id', 'second',
                         'exposure_value']]
```

```python
# add a new column for indicating good video based on recording duration
df_part_useful['well_recorded'] = df_part_useful['duration_recording'].apply(lambda x: 1 if x >= 120 else 0)
```

```python
# fill the rest of recording column with 0
df_part_useful['well_recorded'] = df_part_useful['well_recorded'].fillna(0)
```

```python
# get final dataset
df_final = df_part_useful.drop(['duration_recording'], axis = 1)
```

```python
# send the dataset into csv
df_final.to_csv('data_cleaned.csv')
```
